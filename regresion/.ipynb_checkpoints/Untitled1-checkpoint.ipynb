```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api  as smf
```

```python
df = pd.DataFrame(np.random.rand(1000,5),
                  columns=["x{0}".format(i) for i in range(1,6)])
df["y"] = 6+ 3*df["x1"]+df["x2"]*df["x5"]+np.random.randn(1000)
```

```python
df.head()
```

```python
formulas = ["y ~ x1",
           "y ~ x1 + x2",
           "y ~ x1 + x5",
           "y ~ x1 + x2:x5"]
```

```python
for form in formulas:
    model = smf.ols(formula = form,data=df)
    fit = model.fit()
    print(fit.summary())
    print(sm.stats.anova_lm(fit))
```
