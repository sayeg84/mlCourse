```r
# Exposure to cadmium

options(digits=4)  

getwd()

CADdata <- read.table("cadmium.txt",header=TRUE, sep=" ", dec=".")

head(CADdata); tail(CADdata)
str(CADdata)


#DECLARE  age as numeric
CADdata$age <- as.numeric(CADdata$age)
str(CADdata)

#Create a new variable expo as: 
#expo= 1, not exposed. expo=2, low exposure. expo=3, high exposure.

CADdata$expo[CADdata$group==3]<-1
CADdata$expo[CADdata$group==2]<-2
CADdata$expo[CADdata$group==1]<-3

#DECLARE EXPO AS A FACTOR
CADdata$expo<-as.factor(CADdata$expo)
str(CADdata)
attach(CADdata)

#PLOT THE DATA WITH DIFFERENT COLOURS IN EXPOSURE GROUPS

```

```r

plot(age, vitcap, col=c("green","red","blue")[expo],xlab = "Age", 
     ylab = "Vital Capacity (L)",xlim=c(15,68), ylim=c(2.5,6), 
     pch=c(16,16,16), main = "Exposure to cadmium")

legend(57.2,6.2, c("Not exposed","Low","High" ),
col=c("green","red","blue"), pch = c(16,16,16) )

#TWO Boxplots PLOTS NEXT EACH OTHER
par(mfrow=c(1,2)); par(mar=c(4,4,2,1.5))
boxplot(vitcap ~ expo, data=CADdata, ylab='Vital Capacity (L)', las = 
          1, xaxt = "n",col = 5)
axis(1, at = c(1,2,3), labels = c("Not exposed","Low","High"))

boxplot(age ~ expo, data=CADdata ,ylab='Age',las = 1, xaxt = "n", col = 5)
axis(1, at = c(1,2,3), labels = c("Not exposed","Low","High"))

par(mfrow=c(1,2)); par(mar=c(4,4,2,1.5))
stripchart(vitcap ~ expo, data=CADdata, vertical=TRUE, method="jitter",
           ylab='Vital capacity (L)', cex=1.2,pch=16, las=1, xaxt = "n", col=2:4)
axis(1, at = c(1,2,3), labels = c("Not exposed","Low","High"))

stripchart(age ~ expo, data=CADdata, vertical=TRUE, method="jitter",
           ylab='Age', cex=1.2,pch=16, las=1, xaxt = "n",col=2:4)
axis(1, at = c(1,2,3), labels = c("Not exposed","Low","High"))

#BACK TO ONE PLOT
par(mfrow=c(1,1))

```

```r

# Fit an additive model, no interaction.  Pararel lines
#Reference group parametrization
Model1<-lm(vitcap ~ age + expo, data=CADdata)
anova(Model1)
drop1(Model1, test="F")
summary(Model1)
#Alternative parametrization
Model1B<-lm(vitcap ~ age + expo -1, data=CADdata)
summary(Model1B)
round(confint(Model1B), digits=2)
coef1B <- coef(Model1B)
coef1B

#**************
        # Auxiliary x values for plotting
        #xaux<-seq(from = 18, to = 65, length.out=100)
        xaux<-seq(from = 15, to = 68, length.out=100)
#Plot points in a single plot
par(mfrow=c(1,1))

plot(age, vitcap, col=c("green","red","blue")[expo],xlab = "Age", 
     ylab = "Vital Capacity (L)",xlim=c(15,68), ylim=c(2.5,6), pch=c(16,16,16), main = "Exposure to cadmium")

legend(57.2,6.2, c("Not exposed","Low","High" ),
       col=c("green","red","blue"), pch = c(16,16,16) )

# plot line for first group, expo=1
lines(xaux, coef1B[2] + coef1B[1]*xaux, col="green", type="l",lwd=1)
# plot line for first group, expo=2
lines(xaux, coef1B[3] + coef1B[1]*xaux, col="red", type="l", lwd=1)
# plot line for first group, expo=3
lines(xaux, coef1B[4] + coef1B[1]*xaux, col="blue", type="l", lwd=1)
#end plot
#*******************

```

```r
#Fit model with interaction

Model2<-lm(vitcap ~ age + expo + age:expo, data=CADdata)
anova(Model2)
drop1(Model2, test="F")
summary(Model2)

#Fit model with alternative paramerization to get interceptors 
# and slopes for each line and for plotting. 

Model2C<-lm(vitcap ~ 0  + expo + age:expo, data=CADdata)
summary(Model2C)
round(confint(Model2C), digits=2)


# Residual plots -the same for any parametrization of the same model.

par(mfrow = c(2,2), mgp = c(2,0.7,0), mar = c(3,3,1.5,1))
plot(Model2, which = 1:4)
plot(Model2)

par(mfrow=c(2,1)); par(mar=c(4,4,2,1.5))
plot(age, residuals(Model2), ylab="Residuals");
plot(residuals(Model2), ylab="Residuals");

#Plot the lines for the model with interaction -final.

#Copy coefficients into an auxiliary vector for easy reading. 
coef <- coef(Model2C)
coef
```

```r

# Auxiliary x values for plotting
#xaux<-seq(from = 18, to = 65, length.out=100)
xaux<-seq(from = 15, to = 68, length.out=100)
#Plot points in a single plot
par(mfrow=c(1,1))

plot(age, vitcap, col=c("green","red","blue")[expo],xlab = "Age", 
     ylab = "Vital Capacity (L)",xlim=c(15,68), ylim=c(2.5,6), pch=c(16,16,16), main = "Exposure to cadmium")

legend(57.2,6.2, c("Not exposed","Low","High" ),
       col=c("green","red","blue"), pch = c(16,16,16) )
# plot line for first group, expo=1
lines(xaux, coef[1] + coef[4]*xaux, col="green", type="l",lwd=1)
# plot line for first group, expo=2
lines(xaux, coef[2] + coef[5]*xaux, col="red", type="l", lwd=1)
# plot line for first group, expo=3
lines(xaux, coef[3] + coef[6]*xaux, col="blue", type="l", lwd=1)
#end plot

#Finally printout coeficients and CI rounded to two decimals

round(coef(Model2C), digits=2)
round(confint(Model2C), digits=2)
#End
```
