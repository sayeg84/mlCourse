```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model 
def minSquare(xs,ys):
    xsf = xs.reshape(-1,1)
    reg = linear_model.LinearRegression()
    reg.fit(xsf,ys)
    return reg.coef_, reg.intercept_
def graphicalRegression(xs,ys):
    plt.figure(figsize=(6,4))
    plt.title("Linear regression")
    plt.scatter(xs,ys,label="data",c="C0")
    m,b = minSquare(xs,ys)
    ysfit = [m*x+b for x in xs]
    plt.plot(xs,ysfit,label="fit",c="k")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
xs = range(20)
xs = np.array(xs)
ys = [3*x - 5 + 8*np.random.rand()-4 for x in xs]
graphicalRegression(xs,ys)
```
